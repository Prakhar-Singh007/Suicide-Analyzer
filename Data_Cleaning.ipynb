```python
import os
import pandas as pd
import matplotlib.pyplot as plt 
from keras.preprocessing.text import Tokenizer
import seaborn as sns
```

```python
cleaned_df = pd.read_csv('Suicide_Detection_cleaned.csv', encoding = "utf-8") 
cleaned_df.head()
```

```python
tokenizer = Tokenizer()
text = pd.concat([cleaned_df['clean_msg'] , cleaned_df['clean_msg']],axis=0).astype("str")
tokenizer.fit_on_texts(text)
```

```python
word_freq = pd.DataFrame(tokenizer.word_counts.items(), columns=['word','count']).sort_values(by='count', ascending=False)
```

```python
plt.figure(figsize=(16, 8))
sns.barplot(x='count',y='word',data=word_freq.iloc[:30])
plt.title('Most Frequent Words')
plt.xlabel("Frequency")
plt.ylabel("Word")
plt.show()
```

```python
cleaned_df['clean_msg'] = cleaned_df['clean_msg'].str.replace('filler', '')
```

# Remove rows that do not contain any words

```python
cleaned_df = cleaned_df.dropna(thresh=4)
cleaned_df.head()
```

# Remove outliers in word count of texts

```python
posts_len = [len(x.split()) for x in cleaned_df['clean_msg']]
pd.Series(posts_len).hist(bins=60)
plt.show()
print(pd.Series(posts_len).describe())
```

```python
cleaned_df = cleaned_df[cleaned_df['clean_msg'].apply(lambda x: len(x.split())<=62)]
cleaned_df.reset_index(drop=True, inplace=True)
```

```python
cleaned_df.head()
cleaned_df.shape
```

```python
cleaned_df.to_csv('suicide_detection_final_cleaned.csv', index=False)
```
